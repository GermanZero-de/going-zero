Raumtypen

```python
#import Raumtypen

#columnNames=['gemeindeId',"gem","wohngeb_insgesamt","wohngeb_1","wohngeb_2","wohngeb_3","wohnheime","wohnfläche",
#           "wohn_insgesamt","wohn_1raum","wohn_2raum","wohn_3raum","wohn_4raum","wohn_5raum","wohn_6raum","wohn_7raum","wohn_gr7raum","Unnamed: 17","ags"]

renameDict={'8-stellige AGS':'ags','gem':'id','gemname':'gem'}
             
raumtypen = pd.read_csv('data/Raumtypen.csv',sep=';',header=0,encoding="utf8")

#drop empty column
raumtypen.drop(columns=["Unnamed: 9","Unnamed: 10"],inplace=True)

raumtypen.rename(columns=renameDict,inplace=True)


#put ags as second column
raumtypen=raumtypen[['id','ags','gem','RegioStaR7','gemrs', 'gembev', 'gemfl', 'vbg', 'vbgrs', 'vbgname']]

#drop last 12 columns which are comments
#Einwohner.drop([i+13908 for i in range(33)],axis=0,inplace=True)

#change cell types to string, int or float defined as in dtypes dict
dtypesDict={'id':'string','ags':'string','gem':'string','RegioStaR7':'int','gemrs':'int64', 'gembev':'int', 'gemfl':'float', 'vbg':'int', 'vbgrs':'int', 'vbgname':'string'}
    
#change data types
for key in dtypesDict:
    #print(key)
    raumtypen[key]=raumtypen[key].astype(dtypesDict[key])
    
#make ags column 8 digits
for i in range(len(raumtypen['ags'])):
    if len(raumtypen.iat[i,1]) < 8 :
        raumtypen.iat[i,1]= '0'+ raumtypen.iat[i,1]

#print(raumtypen.dtypes)  

display(raumtypen)
raumtypen.to_pickle('data/Raumtypen.pki')
```

```python


def get_raumtyp(inVal,output='RegioStaR7',inputType='ags'):
    pop=raumtypen[raumtypen[inputType]==inVal][output]
    return pop.values[0]
```

Wohnungen

```python
#import Wohnungen
columnNames=['id',"gem","wohngeb_insgesamt","wohngeb_1","wohngeb_2","wohngeb_3","wohnheime","wohnfläche",
           "wohn_insgesamt","wohn_1raum","wohn_2raum","wohn_3raum","wohn_4raum","wohn_5raum","wohn_6raum","wohn_7raum","wohn_gr7raum","Unnamed: 17","ags"]


Wohnungen = pd.read_csv('data/Wohnungen.csv',names=columnNames,sep=';',header=6,encoding="utf8")

#drop empty column
Wohnungen.drop(columns=["Unnamed: 17"],inplace=True)

#put ags as second column
Wohnungen=Wohnungen[['id','ags',"gem","wohngeb_insgesamt","wohngeb_1","wohngeb_2","wohngeb_3","wohnheime","wohnfläche",
           "wohn_insgesamt","wohn_1raum","wohn_2raum","wohn_3raum","wohn_4raum","wohn_5raum","wohn_6raum","wohn_7raum","wohn_gr7raum"]]

#drop last 12 columns which are comments
Wohnungen.drop([i+13908 for i in range(12)],axis=0,inplace=True)

#change cell types to string, int or float defined as in dtypes dict
   
dtypesDict={'id':'string',"gem":'string',"wohngeb_insgesamt":'int',"wohngeb_1":'int',"wohngeb_2":'int',"wohngeb_3":'int',"wohnheime":'int',"wohnfläche":'float',
            "wohn_insgesamt":'int',"wohn_1raum":'int',"wohn_2raum":'int',"wohn_3raum":'int',"wohn_4raum":'int',"wohn_5raum":'int',"wohn_6raum":'int',"wohn_7raum":'int',"wohn_gr7raum":'int',"ags":'string'}

#workaround for missing values replace all '-' with -1 
Wohnungen.replace('-',-1,inplace=True)
#change data types
for key in dtypesDict:
    Wohnungen[key]=Wohnungen[key].astype(dtypesDict[key])
#and replace all -1 with NaN   
Wohnungen.replace(-1,np.NaN,inplace=True)

#make ags column 8 digits
for i in range(len(Wohnungen['ags'])):
    if len(Wohnungen.iat[i,1]) < 8 :
        Wohnungen.iat[i,1]= '0'+ Wohnungen.iat[i,1]
        
display(Wohnungen)
Wohnungen.to_pickle('data/Wohnungen.pki')
```

```python
def get_wohnung(inVal,output,inputType='ags'):
    pop=Wohnungen[Wohnungen[inputType]==inVal][output]
    return pop.values[0]
```
