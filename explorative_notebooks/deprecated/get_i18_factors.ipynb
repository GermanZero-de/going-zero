```python
import pandas as pd
import numpy as np
```

In this notebook a file with ratios to calculate the i18 outputs of the generator from the CO2_total value.  
For this the current (25.10.2022) results of the generator for germany are used.

```python
df = pd.read_csv(
    "i18_values_germany.csv",
    decimal=",",
    thousands=".",
    sep=";",
)
```

```python
df[["cat", "type"]] = df["description"].str.split(pat=".", expand=True)[[2, 3]]
```

```python
for categorie in df["cat"].unique():
    co2e_total = df[np.logical_and(df["cat"] == categorie, df["type"] == "CO2e_total")]["value"].values[0]
    df.loc[df["cat"] == categorie, "raito_CO2_total"] = df[df["cat"] == categorie]["value"] / co2e_total
```
